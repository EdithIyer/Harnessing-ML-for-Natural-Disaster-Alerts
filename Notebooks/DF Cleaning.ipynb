# Cleaning

```python
import pandas as pd
```

```python
df_thejakartaglobe = pd.read_csv('./the_jakarta_globe.csv')
```

```python
df_thejakartaglobe.shape
```

```python
df_thejakartaglobe['tweet'][1].split('http://')[0].rstrip()
```

```python
for i in range(len(df_thejakartaglobe['tweet'])):
    df_thejakartaglobe['tweet'][i] = df_thejakartaglobe['tweet'][i].split('http://')[0].rstrip().split('https')[0].rstrip().split('pic.twitter')[0].rstrip()
```

```python
df_thejakartaglobe['tweet'][8]
```

```python
df_thejakartaglobe.to_csv('the_jakarta_globe_cleaned.csv', index=False, sep=',')
```
